```python
x = sin(linspace(0, 2*pi)) + sin(linspace(0, 4*pi))/2 + sin(linspace(0, 6*pi))/3 
plt.plot(x)
plt.title('mixed wave')
```

```python
n = len(x)
y = fft.fft(x)/n

plt.plot(sin(linspace(0, 2*pi))*imag(y[2])/imag(y[2]))
plt.plot(sin(linspace(0, 4*pi))*imag(y[3])/imag(y[2]))
plt.plot(sin(linspace(0, 6*pi))*imag(y[4])/imag(y[2]))
plt.title('wave docomposition')
for i in range(2,6):
    yyy=imag(y[i])/imag(y[2])
    print 'wave number', i-1, ':', yyy 
```
