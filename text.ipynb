```python
x = sin(linspace(0,20))/4 + sin(linspace(0,10))/3 + sin(linspace(0,5))/2 + sin(linspace(0,2))
plt.plot(x)
plt.title('mixed wave')
```

```python
n = len(x)
y = fft.fft(x)/n
y = y[range(n/2)]
plt.plot(y)
plt.title('Frequency')
```

```python
back_x = ifft(y)
plt.plot(back_x)
plt.title('now we are back')
```
